```python
%%sql
SELECT COUNT(name) AS above_average
  FROM facts
 WHERE  CAST(Population AS Float)/Area_land > (SELECT CAST(SUM(population) AS FLOAT)/SUM(area_land) AS average_population_density
                                          FROM facts
                                          WHERE name NOT IN ('World','European Union','Antarctica')
                                          AND population NOT NULL
                                          AND area NOT NULL
                                          AND area_land != 0)
   AND population NOT NULL
   AND area NOT NULL
   AND area_land != 0;
```

```python
%%sql
SELECT name,population,area_land,ROUND(CAST(Population AS Float)/Area_land,3) AS population_density
  FROM facts
 WHERE  CAST(Population AS Float)/Area_land > (SELECT CAST(SUM(population) AS FLOAT)/SUM(area_land) AS average_population_density
                                              FROM facts
                                              WHERE name NOT IN ('World','European Union','Antarctica')
                                              AND population NOT NULL
                                              AND area NOT NULL
                                              AND area_land != 0)
   AND population NOT NULL
   AND area NOT NULL
   AND area_land != 0
ORDER BY population_density DESC
 LIMIT 20;
```

- The population is above average
- The area is below average

```python
%%sql
SELECT name, population, area_land, ROUND(CAST(population AS FLOAT)/area_land,2) AS population_density
  FROM facts
 WHERE population >  (SELECT AVG(population)
                      FROM facts
                      WHERE name NOT IN ('World','European Union','Antarctica')
                      AND population NOT NULL
                      AND area NOT NULL
                      AND area_land != 0)
    
   AND area_land  <  (SELECT AVG(area_land)
                      FROM facts
                      WHERE name NOT IN ('World','European Union','Antarctica')
                      AND population NOT NULL
                      AND area NOT NULL
                      AND area_land != 0)
AND population NOT NULL
AND area NOT NULL
AND area_land != 0
ORDER BY population_density DESC;
```

```python
%%sql
SELECT name, population, area_land, ROUND(CAST(population AS FLOAT)/area_land,2) AS population_density
  FROM facts
 WHERE population_density < (SELECT CAST(SUM(population) AS FLOAT)/SUM(area_land)
                                          FROM facts
                                          WHERE name NOT IN ('World','European Union','Antarctica')
                                          AND population NOT NULL
                                          AND area NOT NULL
                                          AND area_land != 0)
   AND population NOT NULL
   AND area NOT NULL
   AND area_land != 0
ORDER BY population_density
LIMIT 20;
```
